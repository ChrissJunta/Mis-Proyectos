# Regresión Lineal Simple
## Horas Trabajadas vs Productos Terminados
### Notebook completo y didáctico

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

plt.style.use("default")
sns.set()
```

## 1.Cargar datos

```python

df = pd.read_excel("datos_produccion_generados.xlsx")
df.head()
```

## 2. Selección de variables

```python

variable_x = "Horas Trabajadas"
variable_y = "Productos Terminados"

X = df[[variable_x]]
Y = df[variable_y]
```

## 3. Entrenar modelo de regresión lineal

```python

modelo = LinearRegression()
modelo.fit(X, Y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print("Ecuación de la recta: Y =", round(pendiente,3), "X +", round(intercepto,3))
```

## 4. Correlación y R²

```python

correlacion = np.corrcoef(df[variable_x], df[variable_y])[0,1]
r2 = r2_score(Y, modelo.predict(X))

print("Coeficiente de correlación:", round(correlacion,3))
print("Coeficiente de determinación R²:", round(r2,3))
```

## 5. Visualización con intervalo de confianza

```python

sns.regplot(x=df[variable_x], y=df[variable_y], ci=95,
            line_kws={"color":"red"}, scatter_kws={"color":"black"})
plt.title("Regresión Lineal con Intervalo de Confianza al 95%")
plt.show()
```

## 6. Predicción

```python

dato_predictor = 15
X_new = pd.DataFrame([dato_predictor], columns=[variable_x])
prediccion = modelo.predict(X_new)

print(f"Si se trabajan {dato_predictor} horas, se estiman {round(prediccion[0],3)} productos terminados.")
```


## 7.Interpretación del modelo

### ✔ Pendiente  
Indica cuántos productos adicionales se generan por cada hora extra trabajada.

### ✔ Intercepto  
Cantidad estimada de productos cuando las horas = 0 (contextual, no siempre útil).

### ✔ Correlación (r)  
- Muy cercano a 1 → relación fuerte positiva  
- Muy cercano a 0 → no hay relación  

### ✔ R²  
Indica qué porcentaje de la variación de la producción se explica por las horas trabajadas.

### ✔ Gráfico  
Muestra la tendencia general, la línea estimada y el intervalo de confianza.
